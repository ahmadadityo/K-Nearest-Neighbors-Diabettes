```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score,classification_report
from sklearn.metrics import accuracy_score
```

```python
DataSet = pd.DataFrame()
DataSet = pd.read_csv("/content/drive/MyDrive/Belajar Machine Learning Python/Belajar Dari Sumber Lain/Data/diabetes.csv")
DataSet.head(10)
```

```python
for column in DataSet.columns[1:-1]:
    DataSet[column].replace(0,np.NaN, inplace=True)
    DataSet[column].fillna(round(DataSet[column].mean(skipna=True)), inplace=True)
DataSet.head(10)
```

```python
X = DataSet.drop("Outcome",axis=1)
y = DataSet["Outcome"]
# Split data into 20% for testing and 80% for training
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)

X
```

```python
neighbor = KNeighborsClassifier(n_neighbors=5,metric="euclidean")
neighbor.fit(X_train,y_train)
```
